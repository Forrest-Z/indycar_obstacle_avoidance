```python
import numpy as np
import math
import matplotlib.pyplot as plt
import sys
import trajectory_planning_helpers as tph
```

```python
# From TUM min-time trajectory optimization, using "rounded rectangle" track
rr_trackfile = './global_racetrajectory_optimization/inputs/tracks/rounded_rectangle.csv'
min_time_traj_file = './global_racetrajectory_optimization/outputs/traj_race_cl.csv'
```

```python
# Ego Vehicle Params
veh_params = {"v_max": 70.0,     # m/s
              "length": 4.7,     # m
              "width": 2.0,      # m
              "mass": 1200.0,    # kg
              "dragcoeff": 0.75, # [kg*m2/m3]
              "curvlim": 0.12,   # rad/m
              "g": 9.81}         # N/kg
```

```python
# Load racetrack
rr_track = np.loadtxt(rr_trackfile, comments='#', delimiter=',')
# Sample data
# x,y centerline: x_m,  y_m,  w_tr_right_m,  w_tr_left_m
rr_track[0:5,:]  
```

```python
# Load min time trajectory, output of TUM program
mt_traj = np.loadtxt(min_time_traj_file, comments='#', delimiter=';')
# Sample data
# s_m; x_m; y_m; psi_rad; kappa_radpm; vx_mps; ax_mps2
mt_traj[0:5,:]  
```

```python
# spline interpolation using TUM tools

# spline smoothing parameters
reg_smooth_opts={"k_reg": 3,
                 "s_reg": 10}
stepsize_opts={"stepsize_prep": 1.0,
               "stepsize_reg": 3.0,
               "stepsize_interp_after_opt": 2.0}

reftrack_interp = tph.spline_approximation. \
    spline_approximation(track=rr_track,
                         k_reg=reg_smooth_opts["k_reg"],
                         s_reg=reg_smooth_opts["s_reg"],
                         stepsize_prep=stepsize_opts["stepsize_prep"],
                         stepsize_reg=stepsize_opts["stepsize_reg"],
                         debug=True)

refpath_interp_cl = np.vstack((reftrack_interp[:, :2], reftrack_interp[0, :2]))

coeffs_x_interp, coeffs_y_interp, a_interp, normvec_normalized_interp = tph.calc_splines.\
        calc_splines(path=refpath_interp_cl,
                     use_dist_scaling=False)
```

```python
refline = reftrack_interp[:,0:2]

# calculate track bounds using centerline and interpolated normal vectors
bound1_interp = refline + (normvec_normalized_interp.T * reftrack_interp[:,2]).T
bound2_interp = refline - (normvec_normalized_interp.T * reftrack_interp[:,3]).T

bound1_interp_cl = np.vstack((bound1_interp[:], bound1_interp[0]))
bound2_interp_cl = np.vstack((bound2_interp[:], bound2_interp[0]))
```

```python
# calculate normal vectors w.r.t psi (heading angle)
normvec_normalized_opt = tph.calc_normal_vectors.\
            calc_normal_vectors(mt_traj[:, 3])

veh_bound1_real = mt_traj[:, 1:3] + normvec_normalized_opt * veh_params["width"] / 2
veh_bound2_real = mt_traj[:, 1:3] - normvec_normalized_opt * veh_params["width"] / 2
```

```python
def perp( a ) :
    b = np.empty_like(a)
    b[0] = -a[1]
    b[1] = a[0]
    return b

def seg_intersect(a1,a2, b1,b2) :
    da = a2-a1
    db = b2-b1
    dp = a1-b1
    dap = perp(da)
    denom = np.dot( dap, db)
    num = np.dot( dap, dp )
    return (num / denom.astype(float))*db + b1

def dist_btwn_points(a, b):
    dx = a[0] - b[0]
    dy = a[1] - b[1]
    return np.sqrt(dx**2 + dy**2)
```

```python
def get_layer_bounds(trajectory: np.ndarray, boundary: np.ndarray) -> float:

    traj_pt = np.zeros(2)
    traj_dir = np.zeros(2) 
    layer_edges = np.zeros((trajectory.shape[0],2))
                       
    for i in range(trajectory.shape[0]):
                       
        traj_pt[0] = trajectory[i,1]
        traj_pt[1] = trajectory[i,2]
        
        #  Heading direction
        psi = trajectory[i,3]       
        traj_dir[0] = math.cos(psi + math.pi/2)
        traj_dir[1] = math.sin(psi + math.pi/2)  
        
        # Direction normal to heading
        norm_dir = perp(traj_dir)
        trajpt_extnorm = traj_pt + norm_dir
        
        # Get dist from current traj pt to each boundary point
        dists = np.sqrt(np.power(boundary[:,0] - traj_pt[0], 2) + np.power(boundary[:,1] - traj_pt[1], 2))

        # Find 2 shortest distance points
        idxs = np.argpartition(dists, 2)
        bpt1 = boundary[idxs[0]]
        bpt2 = boundary[idxs[1]]

        # Find normal vector intersection with edge between 2 boundary points
        int_pt = seg_intersect(bpt1, bpt2, traj_pt, trajpt_extnorm)
        layer_edges[i,0] = int_pt[0]
        layer_edges[i,1] = int_pt[1]

    return layer_edges
    
```

```python
def get_layer_pts(endpoint1, endpoint2, buffer, lat_sep):
    v = np.array([endpoint2[0] - endpoint1[0], endpoint2[1] - endpoint1[1]])
    unitvec = v / np.linalg.norm(v)
    layer_pts = []
    layer_pts.append(endpoint1 + unitvec * buffer)  # first point
    z = endpoint2 - unitvec * buffer                # last point
    
    while True:
        p = layer_pts[-1] + unitvec * lat_sep
        if dist_btwn_points(p, z) < lat_sep:
            break
        layer_pts.append(p)
    
    return layer_pts
```

```python
# discretize track into segments using vectors normal to raceline
sep = 5  # sample every 5 trajectory points, gives 10m separation between layers
layer_traj = mt_traj[0::sep,0:4]
layer_norm = normvec_normalized_opt[0::sep]

# Get track boundary endpoints for each layer
layer_endpoints1 = get_layer_bounds(layer_traj[:], bound1_interp)
layer_endpoints2 = get_layer_bounds(layer_traj[:], bound2_interp)

# For each layer, sample points at a fixed interval on line normal to trajectory
layer_points = []
for i in range(layer_endpoints1.shape[0]):
    lp = get_layer_pts(layer_endpoints1[i], layer_endpoints2[i], 2.0, 0.1)
    layer_points.append(lp)
```
