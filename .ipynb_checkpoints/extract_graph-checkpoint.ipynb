## Local Graph Extraction from Cost-to-Go Map

```python
%load_ext autoreload
%autoreload 2

import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Class to create a new graph node 
class Node(): 

    def __init__(self, coord = None, children = [], cost = None, s = None): 
        self.s = s
        self.children = children
        self.cost = cost
        self.coord = coord
        
    def copy(self):
        newNode = Node(coord = self.coord, children = [], cost = self.cost, s = self.s)
        return newNode
```

```python
def extractLocalGraph(startNode, planHorizon, newHead=None, visited=None, startPos = None, virtualNode = None): 
    '''
    - Extract local graph starting at startNode and extending until 'planHorizon' distance
    in the Frenet space.
    - Creates a copy of data for relevant section of the graph
    - Adds a virtual node to tie all end vertices together
    
    '''
    clone = None
    if newHead == None:    # initial conditions
        newHead = startNode.copy()
        visited = []
        startPos = startNode.s
        virtualNode = Node()   # virtual end node with no additional cost
    if startNode not in visited and startNode.children != []:
        for child in startNode.children: 
            # planning horizon condition
            if child.s - startPos < planHorizon:
                # Backtracking check so new nodes don't get initialized repeatedly
                if clone is None or (clone is not None and clone.coord != child.coord): 
                    clone = child.copy()
                    
                newHead.children.append(clone) 
                extractLocalGraph(child, planHorizon, clone, visited, startPos, virtualNode) # Depth-First recursion
            
            # Add virtual node to tie endpoints together
            ## TODO: fix to not make multiple copies of virt node; pass in as an argument instead
            if newHead.children == []:  
                newHead.children.append(virtualNode)
            
            # add child node to visited list as recursion unwinds
            visited.append(child)
    return newHead
```

```python
def plotGraph(startNode, visited, ax): 
    '''
    Plots edges and nodes of graph
    '''
    # Plot startNode
    if visited == []:
        ax.plot(startNode.coord[0], startNode.coord[1], 'o', color='blue')
    if startNode.children is not None: 
        for i in startNode.children: 
            if startNode not in visited and i.coord is not None:
                ax.plot(i.coord[0], i.coord[1], 'o', color='blue')
                ax.plot([startNode.coord[0], i.coord[0]], [startNode.coord[1], i.coord[1]], color='red')

                if i not in visited: 
                    plotGraph(i, visited, ax) 
                    visited.append(i)
```

```python
## Build Graph Test Case

g = {}
cost_o = 500
s_o = 25
xlen = 15
ylen = 2  # each side, from middle
for i in range(xlen):
    for j in range(-ylen,ylen+1):
        g[(i,j)] = Node((20*i,3*j), [], cost_o - 20*i + 5*j, s_o + 5*i)
for i in range(xlen-1):
    for j in range(-ylen,ylen+1):
        for k in range(-ylen,ylen+1):
            g[(i,j)].children.append(g[(i+1,k)])
  
```

```python
fig, ax = plt.subplots(figsize=(14,6))
plotGraph(g[(0,0)], [], ax)
```

```python
# Extract Local Graph
localGraph = extractLocalGraph(g[(3,-1)], 25)

# Plot extracted graph
fig, ax = plt.subplots(figsize=(14,6))
plotGraph(localGraph, [], ax)
```

```python
# Check that nodes on end layer of local graph are all connected to the same virtual node (with no coordinates)
ida = id(localGraph.children[0].children[0].children[0].children[0].children[0])
idb = id(localGraph.children[0].children[0].children[0].children[1].children[0])
idc = id(localGraph.children[0].children[0].children[0].children[4].children[0])
assert ida == idb == idc
```

```python
def collision_check(pt, obs):
    '''
    Checks for collisions by determining whether point pt is inside circle obs
    pt is a 2-tuple of form (x,y)
    Obs consists of a 3-tuple (x,y,r) to define the circle
    Returns True for no collision; False indicates collision
    '''
    d = ((pt[0] - obs[0])**2 + (pt[1] - obs[1])**2)**0.5
    if d > obs[2]:
        return True
    else:
        return False
    
```
